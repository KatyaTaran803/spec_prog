# Библиотеки

```python
import pandas as pd # нужно для датафреймов
import requests # нужно для того чтоб достать текст из страницы
from io import StringIO # нужно для того чтоб превратить текст в датафрейм
```

```python
# индексы регионов Украины на сайте https://www.star.nesdis.noaa.gov
dct = {2: 'Chernihiv',
 1: 'Cherkasy',
 4: 'Crimea',
 5: "Dnipropetrovs'k",
 6: "Donets'k",
 7: "Ivano-Frankivs'k",
 8: 'Kharkiv',
 9: 'Kherson',
 10: "Khmel'nyts'kyy",
 11: 'Kiev',
 12: 'Kiev City',
 13: 'Kirovohrad',
 14: "Luhans'k",
 15: "L'viv",
 16: 'Mykolayiv',
 17: 'Odessa',
 18: 'Poltava',
 19: 'Rivne',
 20: "Sevastopol'",
 21: 'Sumy',
 22: "Ternopil'",
 23: 'Transcarpathia',
 24: 'Vinnytsya',
 25: 'Volyn',
 26: 'Zaporizhzhya',
 27: 'Zhytomyr'}
```

# Берем данные для всех регионов Украины

```python
# список, которые будет наполняться датафреймами для каждого региона чтоб потом обьединить их в один
data_list = []
for index in dct.keys():
    print(index)
    try:
        url = f"https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={index}&year1=1981&year2=2020&type=Mean"
        data = requests.get(url).text
        df=pd.read_csv(
                   StringIO(data.split('<pre>')[1]), 
                   header=None, 
                   names=['year','week','NDVI', 'BT', 'VCI','TCI','VHI','f']
                  )
        df['region_index'] = index
        df['region_name'] = dct[index]
    except Exception as e:
        print(f"Error: {e}")
        df = pd.DataFrame(columns=['year','week','NDVI', 'BT', 'VCI','TCI','VHI','f', 'region_index','region_name'])
    data_list.append(df)
df = pd.concat(data_list)
    
```

```python
# убираем лишнюю колонку которая парсится случайным образом
df.drop('f', axis=1, inplace=True)
```

```python
# чистим данные
df.dropna(inplace=True)
df=df[df['NDVI']!=-1]
```

# Меняем индексы стран на новые согласно лабораторной

```python
# новые индексы из лабораторной
dct = {24: 'Chernihiv',
 22: 'Cherkasy',
 25: 'Crimea',
 3: "Dnipropetrovs'k",
 4: "Donets'k",
 8: "Ivano-Frankivs'k",
 19: 'Kharkiv',
 20: 'Kherson',
 21: "Khmel'nyts'kyy",
 9: 'Kiev',
 #12: 'Kiev City',
 10: 'Kirovohrad',
 11: "Luhans'k",
 12: "L'viv",
 13: 'Mykolayiv',
 14: 'Odessa',
 15: 'Poltava',
 16: 'Rivne',
 #20: "Sevastopol'",
 17: 'Sumy',
 18: "Ternopil'",
 6: 'Transcarpathia',
 1: 'Vinnytsya',
 2: 'Volyn',
 7: 'Zaporizhzhya',
 5: 'Zhytomyr'}
```

```python
new_dict = dict([(value, key) for key, value in dct.items()]) 
```

```python
new_dict
```

```python
# удаляем те регионы которых нет в документе лабораторной
df = df[~df['region_index'].isin([12,20])]
```

```python
df['new_region_index'] = df['region_name'].apply(lambda x: new_dict.get(x))
```

```python
# Оставляем только VHI
df = df[['year', 'week', 'VHI', 'region_index',
       'region_name', 'new_region_index']]
```

# Реалізувати процедури для формування вибірок

```python
# Максимум VHI для каждого года для каждого региона
df.drop('week', axis=1).groupby(['year', 'region_index', 'region_name', 'new_region_index']).max()
```

```python
# Минимум VHI для каждого года для каждого региона
df.drop('week', axis=1).groupby(['year', 'region_index', 'region_name', 'new_region_index']).min()
```

```python
# Года с экстримальной засухой VHI < 15 для каждого региона
df[df['VHI'] < 15]
```

```python
# Года с умеренной засухой VHI > 15 и VHI < 35 для каждого региона
df[(df['VHI'] > 15)&(df['VHI'] < 35)]
```

```python
# Сохранить исходные данные в файл
df.to_csv('VHI_info.csv', index=False)
```
